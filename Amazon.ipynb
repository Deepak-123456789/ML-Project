```python

best_sellers = amazon_best_sellers.copy()
```

```python
# deleting the unwanted column
del best_sellers['Unnamed: 0']
```

```python

list = [2020, 2021]

# drop rows from the above list
best_sellers = best_sellers[~best_sellers.year.isin(list) == True]
best_sellers
```

```python
# let's look at the percentages of missing values again
misssing_percent= best_sellers.isnull().mean().round(2) * 100
misssing_percent_df = pd.DataFrame(misssing_percent, columns=['missing values'])
misssing_percent_df
```

```python
best_sellers = best_sellers.dropna(subset=['title'])
best_sellers
```

```python
best_sellers.isnull().sum()
```

```python
# let's check the data types of our columns
best_sellers.dtypes
```

```python
best_sellers.ranks = best_sellers.ranks.apply(int)  # round instead of int is you prefer to round
best_sellers
```

```python
best_sellers.info()
```

```python
best_sellers['no_of_reviews'] = best_sellers['no_of_reviews'].astype(int)
```

```python
# Getting the value count of each genre variable in a dataframe
pd.DataFrame(best_sellers['genre'].value_counts())
```

```python
# Getting the percentage of each gender value counts
pd.DataFrame(best_sellers['genre'].value_counts(normalize=True)*100)
```

```python


# Drop duplicates and get genre counts
g_count = best_sellers.drop_duplicates('title')['genre'].value_counts()

fig, ax = plt.subplots(figsize=(8, 8))

genre_col = ['navy', 'crimson']

# Plot a donut-style pie chart
ax.pie(
    g_count.values,
    labels=g_count.index,
    autopct='%1.1f%%',
    startangle=90,
    textprops={'fontsize': 20},
    colors=genre_col,
    wedgeprops={'width': 0.3}  # Creates the "donut" effect
)


plt.title('Distribution of Genre for Unique Books (2009â€“2019)', fontsize=18)

plt.show()
```

```python
years = np.arange(2009, 2020)

# Create a subplot with 2 rows and 6 columns
fig, ax = plt.subplots(2, 6, figsize=(15, 7))


genre_col = ['navy', 'crimson']

# Overall pie chart for all years
g_count = best_sellers['genre'].value_counts()
ax[0, 0].pie(g_count.values, autopct='%1.1f%%', startangle=90, textprops={'size': 12, 'color': 'white'}, colors=genre_col)
ax[0, 0].set_title('2009 - 2019 (Overall)', fontsize=15)

# Pie charts for each year
for i, year in enumerate(years):
    counts = best_sellers[best_sellers['year'] == year]['genre'].value_counts()
    row, col = divmod(i + 1, 6)  # Calculate row and column index for the subplot grid
    ax[row, col].pie(counts.values, autopct='%1.1f%%', startangle=90, textprops={'size': 12, 'color': 'white'}, colors=genre_col)
    ax[row, col].set_title(str(year), fontsize=12)


plt.suptitle('Genre Distribution of Books (2009-2019)', fontsize=20)
plt.tight_layout()
plt.show()
```

```python


# Assuming `best_sellers` is a DataFrame and `cover_type` is a column in it
data = best_sellers['cover_type'].value_counts().reset_index()
data.columns = ['cover_type', 'count']  # Rename columns for clarity

plt.figure(figsize=[8, 7])
ax = sns.barplot(
    data=data,
    y='cover_type',
    x='count',
    color='b'
)


for container in ax.containers:
    ax.bar_label(container)

plt.xlabel('No. of Times', fontsize=20)
plt.ylabel('Cover Type', fontsize=20)
plt.title("Distribution of Cover Type", weight='bold', fontsize=20)
plt.show()
```

```python
# Assuming `best_sellers` is a DataFrame and `author` is a column in it
data = best_sellers['author'].value_counts().head(10).reset_index()
data.columns = ['author', 'count']
plt.figure(figsize=[8, 7])
ax = sns.barplot(
    data=data,
    y='author',
    x='count',
    color='b'
)


for container in ax.containers:
    ax.bar_label(container)

plt.xlabel('No. of Times', fontsize=20)
plt.ylabel('Authors', fontsize=20)
plt.title("Top 10 Most Appeared Authors", weight='bold', fontsize=20)
plt.show()
```
